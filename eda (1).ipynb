```python
import re
parsed_data = []

# Loop through each line and extract relevant information
    # Split the line by commas
    #parts = re.split(line.split(",,"))
delimiters = ', | ,,'
for line in data:
    parts = re.split(delimiters, line)  # Split using the defined delimiters
    # Extract the college name from the first part
    college = parts[0].split(": ")[0].strip()
    
    # Extract GPA values for Freshmen, Sophomores, Juniors, Seniors
    # GPA values are at fixed positions in the split parts
    gpa_values = []
    for i in range(1, len(parts), 2):
        try:
        # Split and strip the parts, then access the second element
            gpa_str = parts[i].split(",")[1].strip()  # Strip any trailing spaces
            gpa = float(gpa_str)  # Convert to float
        except (IndexError, ValueError):
            gpa = None  # Handle missing or invalid values
    gpa_values.append(gpa)

    # Append the college name and GPA values to parsed_data
    parsed_data.append([college] + gpa_values)
#print(parsed_data)
# # Create a DataFrame with grade levels as columns
columns = [ "Freshmen", "Sophomores", "Juniors", "Seniors"]
df = pd.DataFrame(parsed_data, columns=columns)

# # Display the DataFrame
df
```

```python
import csv
import pandas as pd
college_gpa_dict = {}
columns = ['Freshman', 'Sophomore', 'Junior', 'Senior'] 
# Preprocess the file to handle irregularities
with open('gpareport23.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    rows = [row for row in reader if any(cell.strip() for cell in row)]  # Remove empty rows

college_gpa_dict = {}
columns = ['Freshman', 'Sophomore', 'Junior', 'Senior']
# Convert to DataFrame, skipping the first 4 rows of headers
df = pd.DataFrame(rows[4:], columns=rows[3])  # Use the 4th row as column names

# Clean up the DataFrame
df = df.drop(columns=[col for col in df.columns if "unnamed" in col.lower()])  # Drop unnecessary columns
df.columns = df.columns.str.strip().str.lower()  # Standardize column names
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)  # Trim whitespace from all string values

# Populate the dictionary with GPA data
for index, row in df.iterrows():
    college_name = row.iloc[0]  # First column is the college name
    if college_name not in college_gpa_dict:
        college_gpa_dict[college_name] = {}
    
    for i, column in enumerate(columns):
        if i + 1 < len(row):
            gpa_value = row.iloc[i + 1]
            college_gpa_dict[college_name][column] = gpa_value if pd.notna(gpa_value) else None

# Print the final dictionary
college_gpa_dict
```

```python
import re
import pandas as pd
df = pd.read_csv('UpdatedFinalGPAProject - Table 1.csv')
df
cleaned_data = []


# Split content by newline character
data= content.split('\n')
raw_data=data[31:]
# Print the first few lines
for i, line in enumerate(raw_data[:266]):  # Show the first 10 lines
    print(f"Line {i+1}: {line}")


# Iterate through each line
for line in raw_data:
    # Extract the college name and all GPAs using regex
    match = re.match(r"([A-Za-z &]+),.*?([\d\.]+(?:,[\d\.]+)*)", line)
    if match:
        college = match.group(1).strip()  # Extract the college name
        gpas = re.findall(r"\d+\.\d+", line)  # Find all GPAs in the line
        
        # Check if we have exactly 5 GPA values to match the columns
        if len(gpas) == 5:
            cleaned_data.append({
                "College": college,
                "Freshmen": float(gpas[0]),
                "Sophomores": float(gpas[1]),
                "Juniors": float(gpas[2]),
                "Seniors": float(gpas[3]),
                "Total": float(gpas[4])
            })

# Convert cleaned data into a Pandas DataFrame
df = pd.DataFrame(cleaned_data)

# Display the DataFrame
df
```
